# Preparing the Data 

```python
#creating the dependent and independent vectors
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
#splitting the training and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 42)
```

```python
#feature scaling on the independent variables
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Building the Model for Classification

```python
#training the logistic regression model on the training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

```python
#checking for feature importance
feature_imp_df = pd.DataFrame({
    "Features" : list(df.columns[:-1]),
    "coefficients" : classifier.coef_[0]
})
feature_imp_df
```

```python
#predicting the test set values
y_pred = classifier.predict(x_test)
```

```python
#evaluating the model on the test set
from sklearn.metrics import confusion_matrix, accuracy_score
print(f"This is the confusion matrix: \n{confusion_matrix(y_test, y_pred)}")
```

```python
print(f"This is the accuracy of model based on the test set: {accuracy_score(y_test, y_pred)*100}")
```

# K-fold Cross Validation

```python
#better evaluation of the model using k-fold cross validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = x_train, y = y_train, cv = 10, n_jobs = -1)
print(f"The mean accuracy from the 10 test models: {accuracies.mean()*100}")
```

```python
print(f"The Standard Deviation of the accuracies of 10 test models from the mean Accuracy: {accuracies.std()}")
```
